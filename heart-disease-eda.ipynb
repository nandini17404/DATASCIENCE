```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

df = pd.read_csv('/kaggle/input/datasets/ovigupta/heartfailureprediction/heart.csv')
```

```python
sns.displot(df['Age'],kde=True)
```

# **CATEGORICAL COLUMN** # 

```python
df.groupby('Sex').size().plot(kind="pie",autopct="%0.2f",figsize = (6,6))
```

```python
df.groupby('ChestPainType').size().plot(kind="pie",autopct="%0.2f",figsize = (6,6))
```

```python
df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```
